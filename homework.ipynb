### Problem 10

The variables `q10` is initialized with 5 steps of the Collatz sequence starting from 15.
Collatz sequences work with the following rule.

```
if the current value x is odd:
    the next value will be 3x+1
else:
    the next value will be x/2
```

Extend `q10` to ten entries following this rule, then plot them as a line chart.
Use the values in `q10` as the y values.
You can use the default values for x in the `plt.plot` call.

````python
# DO NOT CHANGE

q10 = [15, 46, 23, 70, 35]
````

````python
# Problem 10: extend the Collatz sequence to 10 entries and plot
while len(q10) < 10:
    x = q10[-1]
    if x % 2 == 1:
        q10.append(3*x + 1)
    else:
        q10.append(x // 2)
plt.figure(figsize=(6,4))
plt.plot(q10, marker='o')
plt.title('Problem 10: Collatz sequence starting at 15')
plt.xlabel('step')
plt.ylabel('value')
plt.grid(True)
plt.show()
````

### Problem 17

Plot the factorial function from n=1 to 10 using log scale for better legibility.

Recall that the factorial of $n$, denoted $n!$, is calculated as the product $1 \times 2 \times \ldots \times n$.
The plots that you plot should be $(1, 1!), (2, 2!), (3, 3!), \ldots, (10, 10!)$.
Use [`plt.yscale`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.yscale.html) to change your y axis to use log scale.

````python
# Problem 17: plot factorial from 1 to 10 with log scale on y axis
ns = list(range(1, 11))
facts = [math.factorial(n) for n in ns]
plt.figure(figsize=(6,4))
plt.plot(ns, facts, marker='o')
plt.yscale('log')
plt.title('Problem 17: Factorial (log scale)')
plt.xlabel('n')
plt.ylabel('n! (log scale)')
plt.grid(True, which='both')
plt.show()
````

### Problem 18

Read the data from `f18.tsv` and parse it as a TSV file.
Divide the data by the "animal" column and in a subplot for each animal, plot the "age" and "weight" columns as the x and y axes respectively with a different subplot for each animal.
Each subplot should be a line chart, but add markers for each point as some animals only have one data point.
Make sure to mention the animal name in the chart titles, and label the axes.

````python
# Problem 18: read f18.tsv and plot age vs weight per animal in subplots
import csv
from collections import defaultdict
data = defaultdict(lambda: {'age': [], 'weight': []})
with open('f18.tsv', newline='') as f:
    reader = csv.DictReader(f, delimiter='	')
    for row in reader:
        animal = row['animal']
        data[animal]['age'].append(float(row['age']))
        data[animal]['weight'].append(float(row['weight']))
animals = list(data.keys())
fig, axes = plt.subplots(len(animals), 1, figsize=(6, 3*len(animals)))
if len(animals) == 1:
    axes = [axes]
for ax, animal in zip(axes, animals):
    ax.plot(data[animal]['age'], data[animal]['weight'], marker='o')
    ax.set_title(animal)
    ax.set_xlabel('age')
    ax.set_ylabel('weight')
    ax.grid(True)
plt.tight_layout()
plt.show()
````

Hint: you may find it useful to separate the rows by color first, and use a separate call to `plt.scatter` for each color.

````python
# Problem 20: read f20.tsv and scatter x,y colored by color column
import csv
from collections import defaultdict
groups = defaultdict(lambda: {'x': [], 'y': []})
with open('f20.tsv', newline='') as f:
    reader = csv.DictReader(f, delimiter='	')
    for row in reader:
        c = row['color']
        groups[c]['x'].append(float(row['x']))
        groups[c]['y'].append(float(row['y']))
plt.figure(figsize=(6,6))
for c, vals in groups.items():
    plt.scatter(vals['x'], vals['y'], s=10, label=c)
plt.title('Problem 20: colored scatter from f20.tsv')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
````
